# TRANSFORMACIÓN DE DATOS



```python
# Eliminamos las columnas que no necesitaremos en nuestro análisis.
df_multas.drop(['CODIGO', 'NUMMUL','FECHASISTEMA', 'CODIGOMULTA','DESCRIPCION', 'DEPARTAMENTO',
                'PROVINCIA', 'DISTRITO', 'UBIGEO', "FECHA_CORTE"], axis=1, inplace=True)
```

```python
# Convertimos las columnas involucradas a tipo fecha.
df_multas["FECHAMULTA_IMPUESTA"] = pd.to_datetime(df_multas["FECHAMULTA"], format="%Y%m%d")
df_multas["FECHAMULTA_PAGADA"] = pd.to_datetime(df_multas["FECHAPROYECCION"], format="%Y%m%d")
```

```python
# Eliminamos las columnas que fueron transformadas al tipo fecha.
df_multas.drop(['FECHAMULTA', 'FECHAPROYECCION'], axis=1, inplace=True)
```

```python
# Calculamos la diferencia en días.
df_multas["DIFERENCIA_DIAS"] = (df_multas["FECHAMULTA_PAGADA"] - df_multas["FECHAMULTA_IMPUESTA"]).dt.days
```

```python
# Creamos dos nuevos sub conjuntos con estado pagada y ausente, eliminando los datos negativos.
multas_limpias_pagadas = df_multas[(df_multas["DIFERENCIA_DIAS"] >= 0 ) & ( df_multas["ESTADO"] == "P" )]
multas_limpias_pendientes = df_multas[(df_multas["DIFERENCIA_DIAS"] >= 0 ) & ( df_multas["ESTADO"] == "A" )]
```

```python
# Seleccionamos los 10 tipos de multas mas repetidos
multas_principales_pagadas = multas_limpias_pagadas["GIRO"].value_counts(ascending=False).head(10).index
multas_principales_pendientes = multas_limpias_pendientes["GIRO"].value_counts(ascending=False).head(10).index
```

```python
# Filtramos las multas necesarias para analizar.
multas_filtradas_pagadas = multas_limpias_pagadas[multas_limpias_pagadas["GIRO"].isin(multas_principales_pagadas)]
multas_filtradas_pendientes = multas_limpias_pendientes[multas_limpias_pendientes["GIRO"].isin(multas_principales_pendientes)]
multas_filtradas_rapidez_pago = multas_limpias_pagadas[multas_limpias_pagadas['GIRO'].isin(['COMERCIO INFORMAL', 'TRANSPORTE'])]
```

```python
# Función para dividir un dataframe con y sin descuento.
def dividir_por_descuento(multas_filtradas, columna='DESCUENTO'):
  con_descuento = multas_filtradas[multas_filtradas[columna].notnull() & (multas_filtradas[columna] > 0)]
  sin_descuento = multas_filtradas[multas_filtradas[columna].isnull() | (multas_filtradas[columna] == 0)]
  return con_descuento, sin_descuento
```

```python
# Función para quitar outlaiers de una lista de giros(tipos de multa).
def eliminar_outlaiers_por_giro(multas_filtradas, lista_giros, columna_outlaiers='DIFERENCIA_DIAS'):
  dataframe_sin_outlaiers = pd.DataFrame()
  for i in lista_giros:
    multa = multas_filtradas[multas_filtradas['GIRO']==i]
    # Definimos los cuartiles, el IQR y los limites
    Q1 = multa[columna_outlaiers].quantile(0.25)
    Q3 = multa[columna_outlaiers].quantile(0.75)
    IQR = Q3 - Q1
    inferior = Q1 - 1.5 * IQR
    superior = Q3 + 1.5 * IQR
    # Filtramos y eliminamos los outlaiers del dataframe
    multas_sin_outliers = multa[(multa[columna_outlaiers] >= inferior) & (multa[columna_outlaiers] <= superior)]
    # Concatenar los resultados
    dataframe_sin_outlaiers = pd.concat([dataframe_sin_outlaiers, multas_sin_outliers])
  return dataframe_sin_outlaiers
```
